## Getting Saved Data

```python
#Adjust these if needed; takes a single cell
lat = 38.125
lon = -101.875
```

```python
#month labels
monthsAbrev = ['Jan','Feb', 'Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
monthsFull = ['January','February', 'March','April','May','June','July','August','September','October','November','December']

#gettting raw data
X_train = xr.open_dataset('/glade/work/kwheelan/linear_data/X_train_temp_bymonth.nc')
Y_train = xr.open_dataset('/glade/work/kwheelan/linear_data/Y_train_temp_bymonth.nc')
X_test = xr.open_dataset('/glade/work/kwheelan/linear_data/X_test_temp_bymonth.nc')
Y_test = xr.open_dataset('/glade/work/kwheelan/linear_data/Y_test_temp_bymonth.nc')

#import observations
obs_file = "/glade/p/cisl/risc/narccap/obs/gridMET/common/DCA/tmax.gridMET.NAM-22i.SGP.nc"
obs = xr.open_mfdataset(obs_file).sel(time = slice('1979-01-01','2014-12-31'))
Y_all = obs.sel(lat = lat, lon = lon, method = 'nearest').sel(time = slice('1980-01-01','2014-12-31'))
all_preds = [key for key in Y_all.keys()] #the names of the predictors
Y_all['month'] = Y_all.time.dt.month
if not 'month' in all_preds: all_preds += ['month'] #adding "month" to the list of variable names

#getting predictions
base = '/glade/work/kwheelan/linear_data/finalPreds_{}_tmax' + '_{}_{}.nc'.format(str(lat),str(lon))
X_preds_lin = xr.open_dataset(base.format('lin'))
X_preds_lasso = xr.open_dataset(base.format('lasso'))
X_preds_lasso_annual = xr.open_dataset(base.format('lasso_annual'))
```

## Plotting Average Temperature Max by Month

```python
#set up to filter by month
monthObs, lin, lasso, lasso_annual = Y_all, X_preds_lin, X_preds_lasso, X_preds_lasso_annual
monthObs['time'], lin['time'], lasso['time'], lasso_annual['time'] = monthObs.month, lin.month, lasso.month, lasso_annual.month

#get lists of monthly temp averages
obsAvgs = [float(monthObs.sel(time = m).mean(dim = 'time').tmax) for m in range(1,13)]
linAvgs = [float(X_preds_lin.sel(time = m).mean(dim = 'time').preds) for m in range(1,13)]
lassoAvgs = [float(X_preds_lasso.sel(time = m).mean(dim = 'time').preds) for m in range(1,13)]
lassoAnAvgs = [float(X_preds_lasso_annual.sel(time = m).mean(dim = 'time').preds) for m in range(1,13)]
#bias = [m-o for (m,o) in zip(modelAvgs, obsAvgs)]
```

```python
def plotModels(x,  ys):
    """Plots obs and 3 models for given data"""
    plt.plot(x, ys[0], '-b', label = 'obs')
    plt.plot(x, ys[1], '-r', label = "handpicked")
    plt.plot(x, ys[2], '-g', label = "LASSO monthly")
    plt.plot(x, ys[3], '-y', label = "LASSO annual")
    
plotModels(monthsAbrev, [obsAvgs, linAvgs, lassoAvgs, lassoAnAvgs])
plt.title("Mean Monthly Max Temperature for Observed and Modeled Data")
plt.ylabel('Temperature (Celcius)')
plt.xlabel('Month')
plt.legend()
plt.show()
```

```python
#get lists of monthly temp averages
obsHotDays = [sum(monthObs.sel(time=m).tmax.values > 35) for m in range(1,13)]
linHotDays = [sum(lin.sel(time=m).preds.values > 35) for m in range(1,13)]
lassoHotDays = [sum(lasso.sel(time=m).preds.values > 35) for m in range(1,13)]
lassoAnHotDays = [sum(lasso_annual.sel(time=m).preds.values > 35) for m in range(1,13)]
```

```python
plotModels(monthsAbrev, [obsHotDays, linHotDays, lassoHotDays, lassoAnHotDays])
plt.title("Number of Days above 35 Degrees Celcius")
plt.ylabel('Number of Days')
plt.xlabel('Month')
plt.legend()
plt.show()
```

```python
Y_all = obs.sel(lat = lat, lon = lon, method = 'nearest').sel(time = slice('1980-01-01','2014-12-31'))
base = '/glade/work/kwheelan/linear_data/finalPreds_{}_tmax' + '_{}_{}.nc'.format(str(lat),str(lon))
X_preds_lin = xr.open_dataset(base.format('lin'))
X_preds_lasso = xr.open_dataset(base.format('lasso'))
X_preds_lasso_annual = xr.open_dataset(base.format('lasso_annual'))

def annualSeasonPlot(startDate, endDate, title = "Seasonal Plot"):
    """input: startDate as a string; ex. '12-01'
              endDate as a string; ex. '02-28'
              title as a string; ex. 'Season plot: winter'
        output: a plot
    """
    winter = startDate[0:2] > endDate[0:2]
    startDate, endDate = '-' + startDate, '-' + endDate
    obsWinterAvgs = [Y_all.sel(time = slice(str(y)+startDate, str(y+winter) + endDate)).tmax.mean(dim = 'time') for y in range(1980, 2014) ]
    linWinterAvgs = [X_preds_lin.sel(time = slice(str(y)+startDate, str(y+winter)+endDate)).preds.mean(dim = 'time') for y in range(1980, 2014) ]
    lassoWinterAvgs = [X_preds_lasso.sel(time = slice(str(y)+startDate, str(y+winter)+endDate)).preds.mean(dim = 'time') for y in range(1980, 2014) ]
    lassoAnWinterAvgs = [X_preds_lasso_annual.sel(time = slice(str(y)+startDate, str(y+winter)+endDate)).preds.mean(dim = 'time') for y in range(1980, 2014) ]
    movingAvg = [sum([obsWinterAvgs[t+i] for i in range(5)])/5 for t in range(len(obsWinterAvgs)-4)]
    
    plotModels(range(1980, 2014), [obsWinterAvgs, linWinterAvgs, lassoWinterAvgs, lassoAnWinterAvgs])
    plt.plot(range(1983, 2013), movingAvg, '-k', label = "5-yr moving average")
    plt.title(title)
    plt.ylabel('Temp (Celcius)')
    plt.xlabel('Year')
    plt.legend()
    plt.show()
    
annualSeasonPlot('12-01', '02-28', "Seasonal Plot: Dec-Jan-Feb")
```

```python
annualSeasonPlot('03-01', '05-30', "Seasonal Plot: Mar-Apr-May")
```
